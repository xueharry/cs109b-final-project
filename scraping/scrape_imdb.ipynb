```python
movie_obj_keys = ['title', 'year', 'director', 'cast', 
                  'writer', 'plot', 'rating', 'votes', 
                  'runtimes', 'color info', 'countries', 
                  'languages', 'certificates', 'mpaa', 
                  'composer', 'production companies']
```

```python
imdb_data_parsed = []
counter = 0
for movie in imdb_data:
    movie_dict = {}
    movie_dict["imdb_id"], movie_obj = movie
    for key in movie_obj_keys:
        if key in movie_obj.keys():
            movie_dict[key] = movie_obj[key]
    imdb_data_parsed.append(movie_dict)
    print(counter, end=", ")
    counter += 1
```

```python
for i in tqdm(range(len(imdb_data_parsed))):
    movie_data = imdb_data_parsed[i]
    for key in ["director", "cast", "writer", "production companies"]:
        if key in movie_data:
            imdb_data_parsed[i][key] = [obj["name"] for obj in movie_data[key]]
```

```python
imdb_dict = {}
for data in imdb_data_parsed:
    imdb_dict[data["imdb_id"]] = data
```
