```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

```python
df_train = pd.read_csv("clean_train.csv")
df_test = pd.read_csv("clean_test.csv")
df_val = pd.read_csv("clean_val.csv")
```

```python
train_labels = df_train.ix[:,11:]
val_labels = df_val.ix[:, 11:]
test_labels = df_test.ix[:, 11:]
```

```python
train_labels.head()
```

```python
# sort genre names alphabetically
genre_names = sorted([genre.replace(".", "_") for genre in list(train_labels.columns)])
```

```python
genre_names
```

```python
# replace dots in column names with underscores
for df in [train_labels, test_labels, val_labels]:
    df["Science_Fiction"] = df["Science.Fiction"]
    df["TV_Movie"] = df["TV.Movie"]
    df.drop("Science.Fiction", axis=1, inplace=True)
    df.drop("TV.Movie", axis=1, inplace=True)
```

```python
train_labels
```

```python
train_labels = train_labels[genre_names]
test_labels = test_labels[genre_names]
val_labels = val_labels[genre_names]
```

```python
train_labels.head()
```

```python
df_train = df_train.ix[:, 1:11]
df_test = df_test.ix[:, 1:11]
df_val = df_val.ix[:, 1:11]
```

```python
predictors = list(df_train.columns)
predictors.remove("id")
```

```python
df_train = df_train[["id"] + predictors]
df_test = df_test[["id"] + predictors]
df_val = df_val[["id"] + predictors]
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_val.head()
```

```python
df_train[genre_names] = train_labels
df_val[genre_names] = val_labels
df_test[genre_names] = test_labels
```

```python
df_train_orig = pd.read_csv("clean_train.csv")
```

```python
df_train_orig["Science_Fiction"] = df_train_orig["Science.Fiction"]
df_train_orig["TV_Movie"] = df_train_orig["TV.Movie"]
```

```python
# make sure the ordering of all columns remained the same
for genre in genre_names:
    assert(list(df_train[genre]) == list(df_train_orig[genre]))
```

```python
for df in [df_train, df_val, df_test]:
    df.drop("Foreign", axis=1, inplace=True)
```

```python
genre_names.remove("Foreign")
```

```python
with open("train_val_test_byid.pkl", "rb") as handle:
    train_test_split = pickle.load(handle)
```

```python
# find what ids are missing from imdb data
missing_train = list(set(train_test_split["train"]) - set(df_train["id"]))
len(missing_train)
```

```python
missing_val = list(set(train_test_split["validation"]) - set(df_val["id"]))
len(missing_val)
```

```python
missing_test = list(set(train_test_split["test"]) - set(df_test["id"]))
len(missing_test)
```

```python
len(train_test_split["validation"])
```

```python
new_train = []
for genre_id in train_test_split["train"]:
    if genre_id in list(df_train["id"]):
        new_train.append(genre_id)
```

```python
new_val = []
for genre_id in train_test_split["validation"]:
    if genre_id in list(df_val["id"]):
        new_val.append(genre_id)
```

```python
new_test = []
for genre_id in train_test_split["test"]:
    if genre_id in list(df_test["id"]):
        new_test.append(genre_id)
```

```python
# remove posters for movies with no data
for file_name in os.listdir("posters_split/train"):
    poster_path = "posters_split/train/" + file_name
    # extract poster id from filename
    poster_id = int(file_name.split("-")[-1].split("_")[0])
    # delete file if poster is not in new training data
    if not poster_id in new_train:
        os.remove(poster_path)
```

```python
# remove posters for movies with no data
for file_name in os.listdir("posters_split/test"):
    poster_path = "posters_split/test/" + file_name
    # extract poster id from filename
    poster_id = int(file_name.split("-")[-1].split("_")[0])
    # delete file if poster is not in new test data
    if not poster_id in new_test:
        os.remove(poster_path)
```

```python
# remove posters for movies with no data
for file_name in os.listdir("posters_split/validation"):
    poster_path = "posters_split/validation/" + file_name
    # extract poster id from filename
    poster_id = int(file_name.split("-")[-1].split("_")[0])
    # delete file if poster is not in new test data
    if not poster_id in new_val:
        os.remove(poster_path)
```

```python
train_poster_ids = []
test_poster_ids = []
val_poster_ids = []

# get list of ids of all posters in directories
for file_name in os.listdir("posters_split/train"):
    train_poster_ids.append(int(file_name.split("-")[-1].split("_")[0]))
# get list of ids of all posters in directories
for file_name in os.listdir("posters_split/test"):
    test_poster_ids.append(int(file_name.split("-")[-1].split("_")[0]))
# get list of ids of all posters in directories
for file_name in os.listdir("posters_split/validation"):
    val_poster_ids.append(int(file_name.split("-")[-1].split("_")[0]))
```

```python
assert(train_poster_ids == new_train)
assert(test_poster_ids == new_test)
assert(val_poster_ids == new_val)
```
